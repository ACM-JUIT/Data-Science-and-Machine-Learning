### figure can be considered as single container that contains axes, text and labels. Axes is a bounding box with ticks and labels which will eventually contain the plot element that make visualisation.

## labelling

```python
plt.plot(x,y)
plt.title("Sine Plot")
plt.xlabel("x")
plt.ylabel("sin x")
```

```python
plt.plot(x,y,'--b',label="sin x")        
plt.plot(x,np.cos(x),'-g',label="cos x")
plt.legend()
plt.axis('equal')
```

```python
ax.plot(x,y)
plt.plot()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
y=np.sin(x)
ax.plot(x,y)
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("Sin x")
ax.set_xlim(-1,11)
ax.set_ylim(-2,2)
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
ax=plt.axes()
ax.plot(x,y)
ax.set(title="Sin Wave",xlim=(-1,11),ylim=(-2,2),xlabel="x",ylabel="sin x")
plt.plot()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,'o')     #o is a marker(thick dot)
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x,y,'>')
```

```python
rng=np.random.RandomState(0)
for m in ['o','.',',','^','v','<','>','+','x','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),m,label="marker = {0}".format(m))
plt.legend()
plt.axis([-1,1.5,-1,1.5])
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x,y,'-ok')
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x,y,'-p',color='0.5',markersize=15,linewidth=4,markerfacecolor='white',markeredgecolor='black',markeredgewidth=2)
```

```python
plt.scatter(x,y,marker='o')
```
